```python
import tensorflow as tf

tensor = tf.constant(range(0, 24))
print(tensor)
```

```python
#reshape
tensor1 = tf.reshape(tensor, [3, 8])

print(tensor1)
```

```python
tensor2 = tf.reshape(tensor1, [-1, 4]) #-1은 '아무래도 상관 없다'라는 의미

print(tensor2)
```

```python
tensor3 = tf.reshape(tensor2, [-1])

print(tensor3)
```

```python
tensor4 = tf.reshape(tensor3, [-1, 2, 6])

print(tensor4)
```

```python
tensor5 = tf.reshape(tensor1, [3, 2, 2, 2])

print(tensor5)
```
