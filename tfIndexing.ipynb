```python
import tensorflow as tf

#벡터 정의
vec = tf.constant([10, 20, 30, 40, 50])
print(vec)
```

```python
print(vec[0])
print(vec[-1])
```

```python
print(vec[:3])
```

```python
#행렬 정의
mat = tf.constant([[10, 20, 30],
                   [40, 50, 60]])

print(mat[0, 2])
```

```python
print(mat[0,:])

print(mat[:, 1])

print(mat[:,:])
```

```python
#3차원 텐서 정의
tensor = tf.constant([[[10, 20, 30],
                      [40, 50, 60]],
                      [[-10, -20, -30],
                      [-40, -50, -60]]])

print(tensor)
```

```python
print(tensor[0,:,:])

print(tensor[:, :2, :2])
```
