```python
import tensorflow as tf

#스칼라 정의
a = tf.constant(1)
b = tf.constant(2)
print("a : ", a)
print("b : ", b)
```

```python
print(tf.rank(a))
```

```python
a = tf.cast(a, tf.float32)
b = tf.cast(b, tf.float32)
print(a.dtype)
print(b.dtype)
```

```python
c = tf.math.add(a, b)
print("result : ", c)
print(tf.rank(c))

print(tf.math.add(a, b))
print(tf.math.subtract(a, b))
print(tf.math.multiply(a, b))
print(tf.math.divide(a, b))
```
