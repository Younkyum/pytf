```python
import tensorflow as tf
import numpy as np

#1차원 배열 정리
py_list = [10., 20., 30.]
num_arr = np.array([10.,10., 10.])

vec1 = tf.constant(py_list, dtype=tf.float32)
vec2 = tf.constant(num_arr, dtype=tf.float32)

print("vec1 : ", vec1)
print("vec2 : ", vec2)
```

```python
print(tf.rank(vec1))
print(tf.rank(vec2))
```

```python
add1 = tf.math.add(vec1, vec2)
print("result : ", add1)
print("rank : ", tf.rank(add1)) #element by element adding
```

```python
#same with upper
add2 = vec1 + vec2
print("result : ", add2)
print("rank : ", tf.rank(add2))
```

```python
#reduce_sum => 벡터 구성 원소 전체 합.
print(tf.reduce_sum(vec1))
print(tf.rank(tf.reduce_sum(vec1)))
```

```python
#square => 벡터 구성 원소 각각 제곱
print(tf.math.square(vec1))
print(tf.rank(tf.math.square(vec1)))
```

```python
#파이썬 연산도 사용 가능
print(vec1 + 1)
print(vec1 ** 2)
print(vec1 ** 0.5)
```
